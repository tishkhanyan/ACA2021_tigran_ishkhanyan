```python
from sklearn.metrics import pairwise_distances

import sys
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
class DBscan:
    
    def __init__(self, eps=0.3, MinPts=2):
        self.eps = eps
        self.MinPts = MinPts - 1
    
    def set_pairwise_distances(self, X):
        self.pairwise_distances = pairwise_distances(X)
        self.pairwise_distances[np.where(self.pairwise_distances > self.eps)] = 0
    
    def eps_shrjakayq(self, x):
        idx = np.where(self.X == x)[0][0]
        result = self.pairwise_distances[idx, :]
        result_idxs = np.where(result != 0)
        return result_idxs
    
    def fit(self, X):
        X = np.array(X)
        self.X = X
        self.set_pairwise_distances(X)
        flag_matr = np.zeros(X.shape[0])
        self.clusters = []
        while 0 in flag_matr:
            X_ = X[np.where(flag_matr == 0)]
            x = X_[np.random.choice(len(X_), 1)]
            lst = [x]
            lst[0] = list(lst[0][0])
            idx = 0
            verjnakan = []
            while idx < len(lst):
                x_ = lst[idx]
                idx += 1
                if flag_matr[(X == x_).all(axis=1).nonzero()[0][0]]:
                    continue
                verjnakan += [x_]
                flag_matr[(X == x_).all(axis=1).nonzero()[0][0]] = 1
                new_points = X[self.eps_shrjakayq(x_)]
                if len(new_points) <= self.MinPts:
                    continue
                    
                lst += new_points.tolist()

            self.clusters += [verjnakan]
            
                
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(1000, noise=0.06)

plt.scatter(X[:, 0],X[:, 1])
```

```python
a = DBscan(eps=0.2, MinPts=2)

a.fit(X)
```

```python
len(a.clusters)
```

```python
for k in range(len(a.clusters)):
    b = np.array(a.clusters[k])[:, 0]
    c = np.array(a.clusters[k])[:, 1]
    plt.scatter(b, c)
```
